```python
data = 'three_channel.h5'

y = io.load_evt_info(data)
display(y)

samples = y.samples.unique()[0]
print(samples)

x = io.load_rwf_info(data, samples)
display(x)
```

```python
# convert df to data dictionary
data_dict = {col: x[col].to_numpy() for col in x.columns}
```

```python
# write it
with uproot.recreate("root_test.root") as file:
    file["rwf"] = data_dict
```

```python
# read it back out
with uproot.open("root_test.root") as file:
    tree = file["rwf"]
    df_read = tree.arrays(library="pd")  # Convert back to a DataFrame

print(df_read.head())
```
